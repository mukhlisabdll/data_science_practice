# Visualisasi Grafik Batang dengan Matplotlib

Pada contoh ini, kita akan membuat **grafik batang (bar chart)** untuk membandingkan nilai dari beberapa kategori menggunakan **Matplotlib**. Grafik batang sering digunakan untuk menunjukkan perbandingan data dalam kategori-kategori tertentu.

Langkah-langkah yang dilakukan:
1. **Import Library**: Kita mengimpor library Matplotlib untuk membuat visualisasi.
2. **Menyiapkan Data**:
   - **Kategori**: Label dari setiap kategori (A, B, C, D, E).
   - **Nilai**: Data numerik yang menunjukkan nilai untuk setiap kategori.
3. **Membuat Grafik Batang**: Fungsi `plt.bar()` digunakan untuk membuat bar chart dengan warna batang yang disesuaikan.
4. **Menambahkan Label dan Judul**: Judul grafik dan label untuk sumbu X dan Y ditambahkan untuk memberikan informasi yang jelas.
5. **Menambahkan Grid**: Grid horizontal ditambahkan untuk mempermudah pembacaan nilai pada grafik.

Hasil visualisasi ini akan membantu melihat perbandingan nilai antar kategori secara lebih jelas.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk plot
kategori = ['A', 'B', 'C', 'D', 'E']  # Kategori yang akan dibandingkan
nilai = [80, 70, 90, 85, 75]  # Nilai dari setiap kategori

# Membuat grafik batang
plt.bar(kategori, nilai, color='orange')

# Menambahkan judul dan label
plt.title("Perbandingan Nilai per Kategori")
plt.xlabel("Kategori")
plt.ylabel("Nilai")

# Menampilkan grid untuk kemudahan visualisasi
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Menampilkan grafik
plt.show()
```

# Visualisasi Box Plot dengan Matplotlib

Pada contoh ini, kita akan membuat **box plot** untuk menggambarkan distribusi nilai dari beberapa grup menggunakan **Matplotlib**. Box plot membantu memvisualisasikan statistik penting seperti median, kuartil, dan pencilan (outliers) dalam data.

Langkah-langkah yang dilakukan:
1. **Import Library**: Library Matplotlib digunakan untuk membuat visualisasi.
2. **Menyiapkan Data**: 
   - Data terdiri dari tiga grup (Grup A, Grup B, Grup C), masing-masing memiliki kumpulan nilai.
   - Distribusi nilai untuk setiap grup akan dibandingkan dalam satu plot.
3. **Membuat Box Plot**:
   - Fungsi `plt.boxplot()` digunakan untuk menghasilkan box plot.
   - Warna kotak (box) dan garis median disesuaikan untuk mempermudah pembacaan.
4. **Menambahkan Label dan Judul**:
   - Judul plot diberikan untuk menunjukkan tujuan visualisasi.
   - Label sumbu Y ditambahkan untuk menunjukkan nilai yang dibandingkan.

Hasil visualisasi ini memberikan gambaran statistik tentang distribusi nilai di setiap grup, membantu kita memahami variabilitas dan pusat data secara lebih intuitif.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk box plot
data = {
    "Grup A": [85, 86, 87, 88, 90, 80, 95],
    "Grup B": [75, 78, 80, 85, 88, 90, 93],
    "Grup C": [85, 87, 88, 89, 90, 92, 94],
}

# Membuat box plot
plt.boxplot(data.values(), tick_labels=data.keys(), patch_artist=True,
            boxprops=dict(facecolor="lightblue", color="blue"),
            medianprops=dict(color="orange", linewidth=2))

# Menambahkan judul dan label
plt.title("Distribusi Nilai Antar Grup")
plt.ylabel("Nilai")

# Menampilkan box plot
plt.show()
```

# Visualisasi Heatmap dengan Matplotlib

Pada contoh ini, kita akan membuat sebuah **heatmap sederhana** menggunakan **Matplotlib** dan **NumPy**. Heatmap digunakan untuk memvisualisasikan data dua dimensi, di mana intensitas warna menunjukkan nilai data.

Langkah-langkah yang dilakukan:
1. **Import Library**: Kita mengimpor Matplotlib untuk membuat plot dan NumPy untuk menangani data array.
2. **Menyiapkan Data**: Data yang digunakan berupa array 2D yang berisi angka.
3. **Membuat Heatmap**: Fungsi `imshow()` dari Matplotlib digunakan untuk menggambarkan data dalam bentuk heatmap.
4. **Menambahkan Colorbar**: Colorbar ditambahkan untuk menunjukkan skala nilai dari warna yang ada.
5. **Menambahkan Label dan Judul**: Label sumbu X, Y, dan judul plot ditambahkan untuk menjelaskan visualisasi.

Dengan kode ini, kita dapat memvisualisasikan bagaimana intensitas nilai berubah seiring dengan perubahan posisi di dalam matriks.


```python
# Import library Matplotlib & Numpy
import matplotlib.pyplot as plt
import numpy as np

# Data untuk heatmap
data = np.array([
    [10, 15, 20, 25],
    [30, 35, 40, 45],
    [50, 55, 60, 65],
    [70, 75, 80, 85]
])

# Membuat heatmap
plt.imshow(data, cmap='viridis', interpolation='nearest')

# Menambahkan color bar
plt.colorbar(label="Intensitas Nilai")

# Menambahkan label
plt.title("Heatmap Sederhana")
plt.xlabel("Kolom")
plt.ylabel("Baris")

# Menampilkan heatmap
plt.show()
```

# Visualisasi Histogram dengan Matplotlib

Pada contoh ini, kita akan membuat **histogram** untuk memvisualisasikan distribusi nilai ujian suatu kelas menggunakan **Matplotlib**. Histogram membantu kita melihat bagaimana data terdistribusi dalam kelompok-kelompok nilai tertentu.

Langkah-langkah yang dilakukan:
1. **Import Library**: Library Matplotlib digunakan untuk membuat histogram.
2. **Menyiapkan Data**: 
   - Data terdiri dari nilai-nilai ujian yang diambil dari beberapa siswa.
   - Histogram akan menunjukkan berapa banyak siswa yang berada dalam rentang nilai tertentu.
3. **Membuat Histogram**:
   - Fungsi `plt.hist()` digunakan untuk membuat histogram.
   - Parameter `bins` diatur ke 5 untuk membagi data ke dalam lima kelompok nilai.
   - Warna histogram disesuaikan menjadi ungu dengan garis tepi hitam untuk mempermudah pembacaan.
4. **Menambahkan Label dan Judul**:
   - Judul grafik dan label sumbu ditambahkan untuk memberikan konteks visualisasi.
   - Sumbu X menunjukkan rentang nilai ujian, sedangkan sumbu Y menunjukkan frekuensi (jumlah siswa).

Hasil visualisasi ini membantu kita memahami bagaimana nilai ujian tersebar di dalam kelas, apakah terdapat kecenderungan tertentu atau distribusi data yang merata.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk histogram
nilai_ujian = [55, 60, 65, 70, 75, 80, 85, 90, 95, 60, 65, 70, 75, 80, 85, 70, 75, 80, 85, 90]

# Membuat histogram
plt.hist(nilai_ujian, bins=5, color='purple', edgecolor='black')

# Menambahkan judul dan label
plt.title("Distibusi Nilai Ujian Kelas")
plt.xlabel("Nilai") 
plt.ylabel("Frekuensi")

# Menampilkan grafik
plt.show()
```

# Visualisasi Grafik Garis dengan Matplotlib

Pada contoh ini, kita akan membuat **grafik garis** untuk menunjukkan perubahan suhu seiring waktu menggunakan **Matplotlib**. Grafik garis digunakan untuk memvisualisasikan hubungan atau pola perubahan antara dua variabel.

Langkah-langkah yang dilakukan:
1. **Import Library**: Library Matplotlib digunakan untuk membuat grafik.
2. **Menyiapkan Data**:
   - Variabel **waktu** menunjukkan waktu dalam satuan jam.
   - Variabel **suhu** menunjukkan nilai suhu dalam derajat Celcius pada setiap waktu tertentu.
3. **Membuat Grafik Garis**:
   - Fungsi `plt.plot()` digunakan untuk menggambar grafik garis.
   - Parameter seperti warna garis, bentuk marker, gaya garis, ketebalan garis, dan ukuran marker disesuaikan agar visualisasi lebih jelas.
4. **Menambahkan Label dan Judul**:
   - Judul grafik ditambahkan untuk menjelaskan hubungan yang divisualisasikan.
   - Label sumbu X menunjukkan waktu, sedangkan sumbu Y menunjukkan suhu.
5. **Menambahkan Grid**: Grid membantu mempermudah pembacaan nilai pada grafik.

Hasil visualisasi ini membantu kita memahami bagaimana suhu berubah seiring waktu, dengan menunjukkan pola kenaikan dan penurunan suhu secara jelas.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk plot
waktu = [1, 2, 3, 4, 5]  # Contoh data waktu (jam)
suhu = [30, 32, 34, 33, 31]  # Contoh data suhu (derajat Celcius)

# Membuat grafik garis
plt.plot(waktu, suhu, color='blue', marker='o', linestyle='-', linewidth=2, markersize=6)

# Menambahkan judul dan label
plt.title("Perubahan Suhu Seiring Waktu")
plt.xlabel("Waktu (jam)")
plt.ylabel("Suhu (C)")

# Menampilkan grid untuk kemudahan visualisasi
plt.grid(True)

# Menampilkan grafik
plt.show()
```

# Visualisasi Pie Chart dengan Matplotlib

Pada contoh ini, kita akan membuat **pie chart** untuk menunjukkan distribusi penggunaan waktu sehari-hari menggunakan **Matplotlib**. Pie chart membantu memvisualisasikan proporsi atau persentase dari setiap aktivitas dalam total waktu yang tersedia.

Langkah-langkah yang dilakukan:
1. **Import Library**: Library Matplotlib digunakan untuk membuat pie chart.
2. **Menyiapkan Data**:
   - Variabel **aktivitas** menunjukkan jenis kegiatan seperti tidur, belajar, olahraga, rekreasi, dan coding.
   - Variabel **waktu** menunjukkan jumlah jam yang dihabiskan untuk setiap aktivitas dalam sehari.
3. **Membuat Pie Chart**:
   - Fungsi `plt.pie()` digunakan untuk menggambar pie chart.
   - Parameter `labels` digunakan untuk menampilkan nama aktivitas di setiap bagian pie.
   - Parameter `autopct='%1.1f%%'` digunakan untuk menampilkan persentase dari setiap bagian pie.
   - Parameter `startangle=90` memutar pie chart agar dimulai dari sudut 90 derajat (posisi atas).
4. **Menambahkan Judul**: Judul diberikan untuk menjelaskan konteks pie chart.

Hasil visualisasi ini memberikan gambaran proporsi waktu yang dihabiskan untuk berbagai aktivitas dalam sehari, sehingga memudahkan analisis pola penggunaan waktu.


```python
# Import library Matplotlib 
import matplotlib.pyplot as plt

# Data untuk pie chart
aktivitas = ["Tidur", "Belajar", "Olahraga", "Rekreasi", "Coding"]
waktu = [8, 6, 2, 3, 5]

# Membuat pie chart
plt.pie(waktu, labels=aktivitas, autopct='%1.1f%%', startangle=90)

# Menambahkan judul
plt.title("Distribusi Penggunaan Waktu Sehari")

# Menampilkan pie chart
plt.show()
```

# Visualisasi Scatter Plot dengan Matplotlib

Pada contoh ini, kita akan membuat **scatter plot** untuk memvisualisasikan hubungan antara tinggi badan dan berat badan menggunakan **Matplotlib**. Scatter plot digunakan untuk menunjukkan pola hubungan antara dua variabel numerik.

Langkah-langkah yang dilakukan:
1. **Import Library**: Library Matplotlib digunakan untuk membuat scatter plot.
2. **Menyiapkan Data**:
   - Variabel **tinggi** menunjukkan data tinggi badan dalam satuan cm.
   - Variabel **berat** menunjukkan data berat badan dalam satuan kg.
3. **Membuat Scatter Plot**:
   - Fungsi `plt.scatter()` digunakan untuk membuat scatter plot.
   - Parameter `color` mengatur warna titik menjadi hijau, sementara `edgecolor` menambahkan garis tepi hitam untuk tiap titik.
   - Parameter `s=100` menentukan ukuran titik agar mudah terlihat.
4. **Menambahkan Label dan Judul**:
   - Judul grafik diberikan untuk menjelaskan hubungan yang divisualisasikan.
   - Sumbu X menunjukkan tinggi badan, sedangkan sumbu Y menunjukkan berat badan.
5. **Menambahkan Grid**: Grid ditambahkan untuk membantu pembacaan nilai secara lebih akurat.

Hasil visualisasi ini membantu kita memahami apakah terdapat pola atau tren tertentu antara tinggi badan dan berat badan, seperti apakah berat badan cenderung bertambah seiring dengan tinggi badan.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk scatter plot
tinggi = [150, 155, 160, 165, 170, 175, 180, 185, 190]
berat = [50, 53, 55, 60, 65, 68, 70, 75, 80]

# Membuat scatter plot
plt.scatter(tinggi, berat, color='green', edgecolor='black', s=100)

# Menambahkan judul dan label
plt.title("Hubungan Tinggi Badan dan Berat Badan")
plt.xlabel("Tinggi Badan (cm)")
plt.ylabel("Berat Badan (kg)")

# Menampilkan grid untuk kemudahan visualisasi
plt.grid(True)

# Menampilkan grafik
plt.show()
```

# Visualisasi Stack Plot dengan Matplotlib

Pada contoh ini, kita akan membuat **stack plot** untuk memvisualisasikan distribusi waktu yang dihabiskan untuk berbagai aktivitas selama seminggu menggunakan **Matplotlib**. Stack plot membantu kita memahami komposisi aktivitas yang berubah seiring waktu.

Langkah-langkah yang dilakukan:
1. **Import Library**: Library Matplotlib digunakan untuk membuat stack plot.
2. **Menyiapkan Data**:
   - Variabel **hari** menunjukkan nama-nama hari dalam seminggu.
   - Variabel aktivitas seperti **belajar**, **makan**, **istirahat**, dan **hiburan** merepresentasikan jumlah waktu (dalam jam) yang dihabiskan untuk masing-masing aktivitas setiap harinya.
3. **Membuat Stack Plot**:
   - Fungsi `plt.stackplot()` digunakan untuk menggambar stack plot.
   - Parameter `labels` digunakan untuk menambahkan label pada legenda.
   - Parameter `colors` menentukan warna untuk setiap aktivitas, sehingga visualisasi lebih mudah dibaca.
4. **Menambahkan Judul dan Label**:
   - Judul diberikan untuk menjelaskan konteks visualisasi.
   - Sumbu X menunjukkan hari dalam seminggu, sedangkan sumbu Y menunjukkan jumlah waktu yang dihabiskan (dalam jam).
5. **Menambahkan Legenda**: Legenda ditempatkan di bagian atas kiri untuk memudahkan identifikasi warna dengan aktivitas yang sesuai.

Hasil visualisasi ini memberikan gambaran menyeluruh tentang bagaimana waktu digunakan untuk berbagai aktivitas dari Senin hingga Jumat. Dengan stack plot, kita bisa dengan mudah melihat perubahan proporsi waktu di antara aktivitas-aktivitas tersebut.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk stack plot
hari = ['Senin', 'Selasa', 'Rabu', 'Kamis', 'Jumat']
belajar = [2, 3, 4, 3, 5]
makan = [1, 1, 1, 1, 1]
istirahat = [1, 2, 1, 2, 2]
hiburan = [2, 1, 2, 1, 3]

# Membuat stack plot
plt.stackplot(hari, belajar, makan, istirahat, hiburan,
              labels=['Belajar', 'Makan', 'Istirahat', 'Hiburan'],
              colors=['#66c2a5', '#fc8d62', '#8da0cb', '#e78ac3'])

# Menambahkan judul dan label
plt.title("Distribusi Waktu dalam Seminggu")
plt.xlabel("Hari")
plt.ylabel("Jam")

# Menambahkan legenda
plt.legend(loc='upper left')

# Menampilkan plot
plt.show()
```

# Membuat Visualisasi dengan Subplots Menggunakan Matplotlib

Dalam contoh ini, kita akan membuat **visualisasi multi-plot** menggunakan Matplotlib dengan 2 baris dan 2 kolom subplots. Setiap plot merepresentasikan jenis data dan visualisasi yang berbeda. Pendekatan ini berguna untuk membandingkan beberapa grafik sekaligus dalam satu canvas.

### Langkah-langkah:
1. **Import Library**: 
   - `matplotlib.pyplot` digunakan untuk membuat grafik dan subplots.
2. **Menyiapkan Data**:
   - Variabel `x` merepresentasikan data pada sumbu X.
   - Variabel `y1`, `y2`, `y3`, dan `y4` adalah data pada sumbu Y yang masing-masing memiliki pola berbeda:
     - `y1`: Kuadrat dari `x`.
     - `y2`: Kuadrat terbalik.
     - `y3`: Fungsi linear.
     - `y4`: Sinyal periodik sederhana.
3. **Membuat Subplots**:
   - Fungsi `plt.subplots(2, 2)` digunakan untuk membuat layout grid dengan 2 baris dan 2 kolom.
   - Setiap plot diakses menggunakan indeks `axs[row, col]`.
4. **Detail Setiap Plot**:
   - **Plot 1 (Grafik Kuadrat)**:
     - Digambarkan dengan garis berwarna biru dan marker.
     - Mewakili fungsi kuadrat.
   - **Plot 2 (Grafik Kuadrat Terbalik)**:
     - Digambarkan menggunakan scatter plot berwarna merah.
     - Mewakili kebalikan fungsi kuadrat.
   - **Plot 3 (Grafik Linear)**:
     - Digambarkan menggunakan grafik batang berwarna hijau.
     - Mewakili hubungan linear.
   - **Plot 4 (Grafik Sinyal Periodik)**:
     - Digambarkan dengan garis putus-putus berwarna ungu.
     - Mewakili pola periodik.
5. **Pengaturan dan Penyesuaian**:
   - Judul, label sumbu X dan Y ditambahkan untuk setiap subplot.
   - Fungsi `plt.tight_layout()` digunakan untuk memberikan jarak antar plot agar tidak saling tumpang tindih.

Hasil visualisasi menunjukkan bagaimana berbagai jenis grafik dapat disajikan secara terorganisir menggunakan subplots. Dengan cara ini, kita dapat mempermudah analisis dan membandingkan pola data yang berbeda dalam satu tampilan.


```python
# Import library Matplotlib
import matplotlib.pyplot as plt

# Data untuk plot
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]  # Kuadrat dari x
y2 = [25, 16, 9, 4, 1]  # Kuadrat terbalik
y3 = [2, 4, 6, 8, 10]  # Linear
y4 = [1, 2, 1, 2, 1]  # Sinyal periodik

# Membuat subplots dengan 2 baris dan 2 kolom
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Plot 1: Grafik garis
axs[0, 0].plot(x, y1, color='blue', marker='o')
axs[0, 0].set_title("Grafik Kuadrat")
axs[0, 0].set_xlabel("X")
axs[0, 0].set_ylabel("Y")

# Plot 2: Grafik scatter
axs[0, 1].scatter(x, y2, color='red')
axs[0, 1].set_title("Grafik Kuadrat Terbalik")
axs[0, 1].set_xlabel("X")
axs[0, 1].set_ylabel("Y")

# Plot 3: Grafik batang
axs[1, 0].bar(x, y3, color='green')
axs[1, 0].set_title("Grafik Linear")
axs[1, 0].set_xlabel("X")
axs[1, 0].set_ylabel("Y")

# Plot 4: Grafik garis
axs[1, 1].plot(x, y4, color='purple', linestyle='--')
axs[1, 1].set_title("Grafik Sinyal Periodik")
axs[1, 1].set_xlabel("X")
axs[1, 1].set_ylabel("Y")

# Menambahkan jarak antar plot
plt.tight_layout()

# Menampilkan grafik
plt.show()
```
